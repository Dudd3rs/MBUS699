This next section is some scatter plots and a linear regression on random integers.  Thanks for the assistance in syntax Gina!!!

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
x=np.random.randint(0,100,size=(2,1000))
plt.scatter(x[0],x[1])
y=x[1]-x[0]
x=np.reshape(x[0],(-1,1))
plt.scatter(x,y)
reg=linear_model.LinearRegression()
reg.fit(x,y)
m,b=reg.coef_[0],reg.intercept_
plt.plot([0,99],[m*0+b,m*99+b],c="red",lw=5)

#Thanks for the syntax lesson!!
```

The Chart below represents daily cycles for a continuous Silver casting process. I would like to classify each section of the process (Heat, Cast, Cool, Wait) based off its' rate of change and capture the inflection points, or points of change, and extract cycle times from the index at those points.  Below is an attempt at a bare classification in order to understand the functionality.

```python
from operator import index
#learning some pandas db for classification study
import pandas as pd
import numpy as np
from sklearn import svm

data=pd.read_csv('https://raw.githubusercontent.com/Dudd3rs/MBUS699/main/CR1000XSeries_Table1_20220307.csv',index_col=0,skiprows=0,parse_dates=['TIMESTAMP'])
plt.title('Furnace Temp Deg. F')
data['Fdelta']=data['Furnace_Body_Avg'].diff(periods=1)
plt.plot(data['Furnace_Body_Avg'])
plt.plot(data['Fdelta'])
data['FClass']=np.where(data['Fdelta']>30,'Heat','')
data.loc[data['Fdelta']<-10,'FClass']='Cool'
data.loc[data['Furnace_Body_Avg']>1995,'FClass']='Cast'
Train=data['2022-03-03':'2022-03-05']
Train=Train['Furnace_Body_Avg'],['FClass']
Test=data['2022-03-05':]
Test=Test['Furnace_Body_Avg'],['FClass']
clf=svm.SVC()
clf.fit(Train,Train['Furnace_Body_Avg'])
```
